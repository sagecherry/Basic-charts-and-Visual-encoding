```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd
Days ={
    "Days":["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
    "Study_Hours":[2,3,4,5,6,7,4],
    "Marks":[20,30,40,50,60,70,30],
    "Attendence":[70,70,60,50,70,60,80],
    "sleep_hours":[6,7,8,6,7,8,4]
}
df=pd.DataFrame(Days)
df
```

```python
# Line chart
plt.plot(df["Days"],df["Study_Hours"],marker="o",color="green")

# X-Axis and Y- Axis label
plt.xlabel("Days")
plt.ylabel("Study Hours")

#Title for the chart
plt.title("Study Hour")
```

```python
# Advance line chart
plt.figure(figsize=(10,5))

#For the Y axis
plt.plot(df["Days"],df["Study_Hours"],marker="^",color="green",label ="Study Hours")

#for the X-Axis
plt.plot(df["Days"],df["Marks"],marker=">",color="red",label ="Marks")

plt.title("Study Hours and Marks")
plt.xlabel("Days")
plt.ylabel("Study Hours and Marks")
plt.legend()
```

```python
#Bar -Chart
plt.figure(figsize=(10,5))
plt.bar(df["Days"],df["Attendence"],color="lightblue")
plt.title("Attendence vs Days")
```

```python
# Advanced bar charts :
plt.figure(figsize=(10,5))
bars = plt.bar(df["Days"],df["Marks"],color="lightblue",label="Attendence")

for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,y,str(y),
           ha ="center",va="bottom")

plt.title("Marks vs Days")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.legend()
plt.grid(axis="y")
plt.show()

```

```python
#Histogram
data={'Marks':[50,60,55,70,80,85,90,45,68,72,77,82,88]}
df=pd.DataFrame(data)
plt.hist(df['Marks'],bins=3)
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show
```

```python
#Scatter Plot(Relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'])
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Conditional Scatter Plot
#Example Category
df['Results']=['Fail','Pass','Fail','Pass','Pass','Pass','Fail']
colors=['red' if r=='Fail' else 'green' for r in df['Results']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

#SeaBorn Graph Visualization

```python
#Histogram
sns.histplot(df['Marks'],kde=True)
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
# Dataset - 2

data = {
    'Day' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Wed', 'Thu', 'Sat', 'Sun'],
    'Region' : ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North'],
    'Sales' : [200, 220, 350, 390, 250, 270, 290, 400, 390],
    'Profit' : [20, 45, 60, 70, 35, 50, 20, 60, 65],
    'Customers' : [20, 30, 25, 40, 35, 40, 45, 47, 33],
    'Category' : ['Electronics', 'Clothing', 'Electronics', 'Food', 'Electronics', 'Clothing', 'Food', 'Clothing', 'Electronics']
}

df = pd.DataFrame(data)
df
```

```python
# 1. Line Chart

plt.figure(figsize = (7,7))

plt.plot(df['Profit'] , df['Day'], marker = 'o', color = 'blue', label = 'Study_Hours')

plt.plot(df['Profit'] , df['Region'], marker = 's', color = 'red', label = 'Marks')

plt.title('Trends of Profit in different days and regions')
plt.xlabel('Profit')
plt.ylabel('Day / Region')
plt.legend()

plt.show()
```

```python
# 2. Bar Graph

bars = plt.bar(df['Category'], df['Sales'], color = 'bisque')

for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, y, str(y), ha = 'center', va = 'bottom')

plt.title("Category wise Sales Analysis")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.grid(axis = 'y') # Adds Horizontal grid for readability

plt.show()
```

```python
# 2.1 Bar Graph

bars = plt.bar(df['Region'], df['Sales'], color = 'paleturquoise')

for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, y, str(y), ha = 'center', va = 'bottom')

plt.title("Category wise Sales Analysis")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.grid(axis = 'y') # Adds Horizontal grid for readability

plt.show()
```

```python
# 3.Histogram

data = {
    'Sales' : [50, 60, 55, 70, 65, 48, 99, 20, 30, 30, 75, 80, 85, 90, 45,20,  68, 72, 77, 82, 88, 67, 90, 95, 67]
}

df1 = pd.DataFrame(data)

plt.hist(df1['Sales'] , bins = 15, color = 'paleturquoise')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Scatter Plot (Relationship)

import matplotlib.pyplot as plt

plt.scatter(df['Region'], df['Sales'], color = 'slateblue')

plt.title("Region vs Sales")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```
